```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el archivo CSV en un DataFrame
yutu = pd.read_csv('MXvideos_cc50_202101.csv')
```

```python
# Verificar qué columnas tienen datos faltantes y cuántos hay en cada una
datos_faltantes = yutu.isnull().sum()
print(datos_faltantes)
```

```python
# Eliminar filas con datos faltantes en columnas críticas
yutu.dropna(subset=['video_id'], inplace=True)
```

```python
# Convertir las columnas necesarias a los tipos adecuados
yutu['trending_date'] = pd.to_datetime(yutu['trending_date'], format='%y.%d.%m', errors='coerce')
yutu['publish_time'] = pd.to_datetime(yutu['publish_time'], errors='coerce')

# Convertir columnas numéricas
numerical_cols = ['views', 'likes', 'dislikes', 'comment_count', 'category_id', 'lat', 'lon']
for col in numerical_cols:
    yutu[col] = pd.to_numeric(yutu[col], errors='coerce')

# Eliminar filas con fechas no válidas
yutu = yutu.dropna(subset=['trending_date', 'publish_time'])

# Reemplazar valores faltantes en columnas categóricas con la moda
yutu['description'].fillna(yutu['description'].mode()[0], inplace=True)
```

```python
# Guardar el DataFrame limpio en un nuevo archivo CSV
yutu.to_csv('Datos_Limpios.csv', index=False)
```
